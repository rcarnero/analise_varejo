```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Análise pelas horas do dia

```python
hour = data['hour'].value_counts().sort_index()

hour_visits = pd.DataFrame(list(zip(hour.index.tolist(), hour.tolist())), 
                                   columns = ['hora', 'visitas_realizadas'])   
hour_visits
```

```python
fig, ax = plt.subplots(1,1,figsize=(20,5))

sns.barplot(x = hour_visits["hora"], y = hour_visits["visitas_realizadas"], palette = sns.color_palette("Blues",1), ax=ax)

ax.set_title('Número de visitas realizadas por hora')
plt.show(fig)


```

Pode-se ver que a movimentação de pessoas pelos estabelecimentos começa a crescer a partir das 6 da manhã e a decrescer a partir das 19, havendo um pico entre as 12 e 13 horas, porém provavelmente a maioria dos estabelecimentos visitados durante essa hora é de restaurante

```python

for curr_hour in range(0, 24):
    categories = {};
    print();
    print(curr_hour,"HORAS:");
    print()
    for i in range(len(data)):
        #print(int(data["hour"][i]), curr_hour)
        if(int(data["hour"][i]) == curr_hour):
            if(not (data["categoria"][i] in categories)): 
                categories[data["categoria"][i]] = 0;
            categories[data["categoria"][i]] += 1;
    category_hour = pd.DataFrame(list(zip(categories.keys(), categories.values())), columns=['Categoria', 'Visitas_realizadas']);
    category_hour = category_hour.sort_values(by='Visitas_realizadas', axis=0, ascending=False).reset_index(drop=True);
    print(category_hour)
    
```

Grocery or supermarket tem o seu pior número de visitas às 18 horas, portanto essa não seria uma boa hora para se fazer anúncios dessa categoria.

```python
def visits_times(data, time):
    value_size = data.groupby([time, 'categoria']).size()
    source_list = value_size.copy().index.tolist()
    time_type, kind = zip(*source_list)
    return list(zip(time_type, kind, value_size.tolist()))

category_hour = pd.DataFrame(visits_times(data, 'hour'), columns = ['hora', 'categoria', 'visitas_realizadas'])
category_hour = category_hour.sort_values(['hora'], ascending = True).reset_index(drop = True)
category_hour
```

```python
def mostCommonVisit(data, init):
    time = init
    num_visits1 = 0
    num_visits2 = 0
    num_visits3 = 0
    visits1 = ""
    visits2 = ""
    visits3 = ""
    most_common_visits = []
    
    for row in data.itertuples(index = False):
        if time != row[0]:
            most_common_visits.append((time, visits1, num_visits1, visits2, num_visits2, visits3, num_visits3))
            num_visits1 = 0
            num_visits2 = 0
            num_visits3 = 0
            time = row[0]
        if row[2] > num_visits1:
            num_visits3 = num_visits2
            num_visits2 = num_visits1
            num_visits1 = row[2]
            visits3 = visits2
            visits2 = visits1
            visits1 = row[1]
        elif row[2] > num_visits2:
            num_visits3 = num_visits2
            num_visits2 = row[2]
            visits3 = visits2
            visits2 = row[1]
        elif row[2] > num_visits3:
            num_visits3 = row[2]
            visits3 = row[1]
            
    most_common_visits.append((time, visits1, num_visits1, visits2, num_visits2, visits3, num_visits3))
    
    return most_common_visits

most_common_category_hour = pd.DataFrame(mostCommonVisit(category_hour, 0), columns = ['Hora', 'Mais Visitado', 'Visitas', 'Segundo mais visitado', 'Visitas', 'Terceiro mais visitado', 'Visitas'])
most_common_category_hour
```

Vemos que Restaurantes e Lojas no geral sempre são os estabelecimentos mais visitados, enquanto que escola é quase sempre o terceiro local mais visitado, perdendo apenas em dois horários. Vemos também que entre 12 e 13 horas todas as 3 melhores categorias tem um aumento significativo no número de visitas em relação às 11 horas.

# Perfil dos clientes

```python
data
mondays = data[data["day_of_week"] == 'Monday'][:].copy()
mondays
```

```python
perfil = {}
avgtime_perfil = {}
mediantime_perfil = {}
usuarios = {}
std = []

for row in mondays.itertuples(index = False):
    if((row[0] in usuarios) and (row[6] in usuarios[row[0]])):
        continue
    if(not (row[0] in usuarios)):
        usuarios[row[0]] = {}
    if(not (row[6] in usuarios[row[0]])):
        usuarios[row[0]][row[6]] = 1
        
    if(not (row[6] in perfil)):
        perfil[row[6]] = 0
        avgtime_perfil[row[6]] = 0
        meantime_perfil[row[6]] = []
    meantime_perfil[row[6]].append(int(row[5]))
    avgtime_perfil[row[6]] += int(row[5])
    perfil[row[6]] += 1

for key in meantime_perfil:
    std.append(np.std(meantime_perfil[key]))
    meantime_perfil[key] = np.median(meantime_perfil[key])
    

perfil_df = pd.DataFrame(list(zip(perfil.keys(), perfil.values(), avgtime_perfil.values(), std, meantime_perfil.values())), 
                         columns = ['categoria', 'visitas_realizadas', 'media_de_hora', 'variacao', 'mediana_da_hora'])
perfil_df = perfil_df.sort_values(by='visitas_realizadas', axis=0, ascending=False).reset_index(drop=True)

perfil_df['media_de_hora'] /= perfil_df['visitas_realizadas']

perfil_df
```
